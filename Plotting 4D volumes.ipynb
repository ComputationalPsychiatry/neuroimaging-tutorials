```python
#http://nilearn.github.io/auto_examples/plot_3d_and_4d_niimg.html#sphx-glr-auto-examples-plot-3d-and-4d-niimg-py

from nilearn import image
from nilearn import plotting
import nibabel as nib

rsn = nib.load("C:/Users/Ege/Desktop/Nilearn/sub001/BOLD/task001_run001/bold.nii.gz")
print(image.load_img(rsn).shape)

#rsn = image.load_img("C:\Users/Ege/Desktop/Nilearn/sub001/BOLD/task001_run001/bold.nii.gz")
#print(rsn.shape)
```

```python
first_rsn = image.index_img(rsn, 0)
print(first_rsn.shape)
```

```python
plotting.plot_stat_map(first_rsn)
plotting.show()
```

```python
plotting.plot_stat_map(first_rsn, threshold = 750)
plotting.show()
```
